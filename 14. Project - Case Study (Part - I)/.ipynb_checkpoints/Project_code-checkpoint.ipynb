```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import dateutil.parser as parser
```

```python
data = pd.read_csv('startup_funding.csv')
d = data.copy()
```

```python
d.head()
```

```python
# Dealing with NaN value

d.fillna('0', inplace = True) #  replace every NaN values with 0
```

```python
# There are two index columns..need to drop one
d.drop("SNo", axis=1, inplace=True)
```

```python
d
```

# TASK 1 : 
1. Find out what type of startups are getting funded over the past few years.<br>
2. Who are important investors.<br>
3. What are the hot fields that get a lot of fundings these days.<br>

```python
# To find types of startups

startups = d[d['IndustryVertical'] != '0']
startups = startups['IndustryVertical']
startups = startups.unique()
startups = pd.DataFrame(startups, columns = ["Name"])
# list of type of startups
startups
```

```python
# Who are important investors.


investors = d[~d["InvestorsName"].str.contains("disclosed" , na=False)]
investors = investors[investors["InvestorsName"]!='0']
investors = investors["InvestorsName"]

investors_names = investors.value_counts()

#top 20 investors
investors_names = investors_names.head(20)


indices = investors_names.index
values = investors_names.values
investors = zip(indices, values)
investors = pd.DataFrame(investors, columns = ['Investors_names', 'Total number of investments'])
investors
```

# TASK 2 -> Check the trend of investments over the year.
1. Total number of funds done in each year
2. Line graph between year and number of fundings. Take year on x-axis and number of fundings on y-axis
3. Print yearwise total number of fundings.
4. There is some error in date feature. Make sure to handle that.

```python
# d['Date'] = d['Date'].dt.strftime('%m/%d/%Y')
# d['Date'] = pd.to_datetime(d.Date, format='%d/%m/%Y')
d['Date'] = pd.to_datetime(d['Date'], errors='coerce')
```

```python
d['Date'] = d['Date'].dt.strftime('%Y')
```

```python
d.sort_values(by=['Date'], inplace = True)
years = d['Date']

years = years.value_counts(ascending=True)
years
```

```python
num = years.values
year  = years.index
```

```python
num = np.array(num)
year = np.array(year)
plt.plot(year, num)
plt.scatter(year, num)
plt.xlabel('Year', fontsize = 14, color = "red")
plt.ylabel("Number of fundings", fontsize = 14, color = "red")
plt.title("Year vs Number of Fundings", fontsize = 15)
```

```python
# table for year and fundings
t = zip(year, num)
t = pd.DataFrame(t, columns = ['Year', 'Total number of fundings'])
t.sort_values(by=['Year'], inplace = True)
t
```

# TASK 3 :
    1. Find out which cities are generally chosen for starting a startup
    2. Find top 10 cities having most number of startups.
    3. Plot a pie chart and visualise it.

```python
d = data.copy()
```

```python
# Find out which cities are generally chosen for starting a startup
cities = d['CityLocation'].str.capitalize()
cities.dropna(inplace = True)
cities = cities.value_counts()

# Find top 10 cities having most number of startups.
cities = cities.head(10)
cities_names = cities.index

#  Plot a pie chart and visualise it.
plt.pie(cities.head(10), radius = 4, autopct = "%.2f%%", labels = cities.index)
plt.show()

# Print cities in descending order with respect to the number of startups
num_startups = cities.values
output = zip(cities_names, num_startups)
p = pd.DataFrame(output, columns = ['City', 'Total no. of startups'])
p
```

# TASK 4: <br>
Find the top 5 startups who received the greatest number of funding rounds<br> i.e. The startups which got fundings maximum number of times
<br>Print the startup name in descending order with respect to the number of <br>funding round as integer value. 

```python
d = data.copy()
d = d[['StartupName']].copy()
```

```python
# replacing the wrong names
d['StartupName'].replace("Ola Cabs", "Ola", inplace = True)
d['StartupName'].replace("Olacabs", "Ola", inplace = True)
d['StartupName'].replace("Flipkart.com", "Flipkart", inplace = True)
d['StartupName'].replace("Oyo Rooms", "Oyo", inplace = True)
d['StartupName'].replace("Oyorooms", "Oyo", inplace = True)
d['StartupName'].replace("OyoRooms", "Oyo", inplace = True)
d['StartupName'].replace("Paytm Marketplace", "Paytm", inplace = True)

# Used str.contains()  function to find out the wrong names and then replaced them.
# paytm = d[d['StartupName'].str.contains('Paytm')]
# paytm
```

```python
# count of funding rounds of all the startups in sorted order
d = d['StartupName'].value_counts()
d
```

```python
# only top 5 startups are to be fetched which are funded maximum number of times
d = d.head(5)
startup = d.index
num_fund = d.values
t = zip(startup, num_fund)
tabl = pd.DataFrame(t, columns = ['Startup', 'No of funding rounds'])
tabl
```

```python
# This pie chart shows the 5 startups and the percentage of how many times each startup received fundings 
explode = [0.1, 0, 0, 0, 0]
plt.pie(d.values, labels = d.index, autopct = "%.1f%%", counterclock = False, startangle = 100, explode = explode)
plt.title("5 startups who received the most number of funding rounds")
plt.show()
```

```python
plt.bar(d.index, d.values, width = 0.59)
plt.xticks(rotation = 40)
plt.title("5 startups who received the most number of funding rounds")
plt.show()
```
